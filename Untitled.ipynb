```python
# Campus Cafeteria Satisfaction Analysis
# Step-wise Data Mining (EDA  Preprocessing  Visualization)
# NO Algorithms  Insight-driven Analysis

# =========================================================
# 1. IMPORT LIBRARIES
# =========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# 2. LOAD DATASET
# =========================================================
df = pd.read_csv("C:/Users/tahap/Downloads/university_cafeteria_survey_dataset1.csv")

# =========================================================
# 3. EXPLORATORY DATA ANALYSIS (EDA)
# =========================================================

# 3.1 Dataset Preview
print(df.head())

# 3.2 Dataset Information
print(df.info())

# 3.3 Statistical Summary
print(df.describe())

# 3.4 Missing Values Check
print(df.isnull().sum())

# =========================================================
# 4. DATA PREPROCESSING
# =========================================================

# 4.1 Handling Missing Values (Numeric)
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# 4.2 Handling Missing Values (Categorical)
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna('Not Specified')

# 4.3 Data Consistency Check
print(df.isnull().sum())

# =========================================================
# 5. DATA VISUALIZATION & PATTERN DISCOVERY
# =========================================================

ratings = ['Food_Taste','Hygiene','Pricing','Waiting_Time','Meal_Variety','Staff_Behavior']

# 5.1 Overall Satisfaction Distribution
plt.figure()
sns.countplot(x='Overall_Satisfaction', data=df)
plt.title('Overall Satisfaction Distribution')
plt.show()

# 5.2 Average Rating of Cafeteria Services
mean_ratings = df[ratings].mean()
plt.figure()
mean_ratings.plot(kind='bar')
plt.title('Average Ratings of Cafeteria Features')
plt.ylabel('Average Score (15)')
plt.show()

# 5.3 Satisfaction vs Individual Features
for col in ratings:
    plt.figure()
    sns.boxplot(x=df[col], y=df['Overall_Satisfaction'])
    plt.title(f'Overall Satisfaction vs {col}')
    plt.show()

# 5.4 Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[ratings + ['Overall_Satisfaction']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Features and Overall Satisfaction')
plt.show()

# 5.5 Gender-wise Satisfaction
plt.figure()
sns.barplot(x='Gender', y='Overall_Satisfaction', data=df)
plt.title('Gender-wise Overall Satisfaction')
plt.show()

# 5.6 Department-wise Satisfaction (Top 10)
top_departments = df['Department'].value_counts().head(10).index
plt.figure(figsize=(10,5))
sns.barplot(x='Department', y='Overall_Satisfaction', data=df[df['Department'].isin(top_departments)])
plt.xticks(rotation=45)
plt.title('Department-wise Satisfaction (Top 10)')
plt.show()

# 5.7 Age vs Overall Satisfaction
plt.figure()
sns.scatterplot(x='Age', y='Overall_Satisfaction', data=df)
plt.title('Age vs Overall Satisfaction')
plt.show()

# =========================================================
# 6. INSIGHTS & OUTCOMES
# =========================================================
print("DATA MINING OUTCOMES:\n")
print("1. Hygiene and Food Taste have the strongest impact on satisfaction.")
print("2. Long waiting time leads to lower satisfaction scores.")
print("3. Pricing dissatisfaction exists among a segment of students.")
print("4. Satisfaction varies across departments and student groups.")
print("5. Key improvement areas: hygiene, food quality, and service speed.")
```
